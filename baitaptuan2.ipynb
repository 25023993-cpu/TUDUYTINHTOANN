Bài 3 : Tính tam giác: Nhâp vào ba số a,b,c bất kì. Kiểmtra xem nó có thể làđộ dài ba cạnh hay không, nếu không thì in ra màn hình 'Khong la ba canh cua mot tam giac'. Ngược lại, kiểm tra xem nó là tam giác gì: Đều, Cân, Vuông, hay Tam giác thường. Tính diện tích, chu vi của tam giác và in ra mànhình

```python
a = float(input())
b = float(input())
c = float(input())
if a + b > c and b + c > a and a + c > b:
    chu_vi = a + b + c
    p = (chu_vi) / 2
    s = (p * (p - a) * (p - b) * (p - c)) ** 0.5
    if a == b == c:
        loai = "La tam giac deu"
    elif a == b or a == c or b == c:
        loai = "La tam giac can"
    elif a * a + b * b == c * c:
        loai = "La tam giac vuong"
    else:
        loai = "La tam giac thuong"
    print(f"Chu vi: {chu_vi}")
    print(f"Dien tich: {s}")
else:
    loai = "Khong phai tam giac"
print(f"Day {loai}")
```

Bài 4: Viết chương trình giải phương trình bậc 1 (ax + b = 0)

```python
a = float(input())
b = float(input())
if a == 0:
    if b == 0:
        print("He co vo so nghiem")
    else:
        print("He vo nghiem")
else:
    x = -b / a
    print(f"He co nghiem duy nhat {x:.0f}")
```

Bài 5: Viết chương trình giải phương trình bậc 2 (ax2 + bx + c = 0)

```python
a = float(input())
b = float(input())
c = float(input())
if a == 0:
    if b == 0:
        if c == 0:
            print("Phuong trinh vo so nghiem")
        else:
            print("Phuong trinh vo nghiem")
    else:
        x = -c / b
        print(f"Phuong trinh co nghiem x = {x:.2f}")
else:
    delta = b * b - 4 * a * c
    if delta < 0:
        print("Phuong trinh vo nghiem")
    elif delta == 0:
        x = - b / (2 * a)
        print(f"Phuong trinh co nghiem x = {x:.2f}")
    else:
        x1 = (- b + pow(delta, 0.5)) / (2 * a)
        x2 = (- b - pow(delta, 0.5)) / (2 * a)
        print("Phuong trinh co 2 nghiem phan biet")
        print(f"x1 = {x1:.2f}")
        print(f"x2 = {x2:.2f}")
```

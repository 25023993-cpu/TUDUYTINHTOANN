```python
lst = list(map(int, input().split()))

def is_beauti(lst):
    for i in lst:
        if lst.count(i) != i:
            return False
    return True

if is_beauti(lst) :
    print(f"day la mang dep")
else :
    new_lst = []
    for i in lst :
        if lst.count(i) == i :
            new_lst.append(i)
    if is_beauti(new_lst):
        print(f"mang dep moi tao ra la {new_lst}")
    if len(new_lst) == 0 :
        print(f"khong the tao mang dep")
```

```python
# Nhập 2 mảng
a = list(map(int, input("Nhap mang a: ").split()))
b = list(map(int, input("Nhap mang b: ").split()))

result = []

i = 0
j = 0

# Kiểm tra tăng hay giảm
tang_dan = True

if len(a) >= 2:
    if a[0] > a[1]:
        tang_dan = False

# Merge
while i < len(a) and j < len(b):

    if tang_dan:

        # Tăng dần
        if a[i] <= b[j]:
            result.append(a[i])
            i += 1
        else:
            result.append(b[j])
            j += 1

    else:

        # Giảm dần
        if a[i] >= b[j]:
            result.append(a[i])
            i += 1
        else:
            result.append(b[j])
            j += 1

# Thêm phần còn lại của a
while i < len(a):
    result.append(a[i])
    i += 1

# Thêm phần còn lại của b
while j < len(b):
    result.append(b[j])
    j += 1

print("Mang sau merge:")
print(*result)
```

```python
n = int(input("Nhap so sinh vien: "))

names = []

# Nhập danh sách tên
for i in range(n):
    full_name = input()

    # Lấy tên cuối cùng
    last_name = full_name.split()[-1]

    names.append(last_name)

# Tìm tên xuất hiện nhiều nhất
max_count = 0
result = ""

for name in names:

    count = names.count(name)

    # Nếu xuất hiện nhiều hơn
    if count > max_count:
        max_count = count
        result = name

    # Nếu bằng nhau -> lấy tên nhỏ hơn theo từ điển
    elif count == max_count:

        if name < result:
            result = name

print("Ten xuat hien nhieu nhat la:", result)
```

```python
number = input("Nhap so: ")
k = int(input("Nhap k: "))

stack = []

remove_count = len(number) - k

for digit in number:

    # Nếu chữ số hiện tại lớn hơn số cuối stack
    # thì xóa số nhỏ hơn để số lớn hơn
    while (remove_count > 0 and
           len(stack) > 0 and
           stack[-1] < digit):

        stack.pop()
        remove_count -= 1

    stack.append(digit)

# Nếu vẫn còn cần xóa
while remove_count > 0:
    stack.pop()
    remove_count -= 1

# Lấy đúng k chữ số
result = "".join(stack[:k])

print("So lon nhat la:", result)
```

```python
s = input("Nhap day ngoac: ")

stack = []

valid = True

for ch in s:

    # Nếu là ngoặc mở
    if ch == '(' or ch == '[' or ch == '{':
        stack.append(ch)

    # Nếu là ngoặc đóng
    else:

        # Stack rỗng -> sai
        if len(stack) == 0:
            valid = False
            break

        top = stack[-1]

        # Kiểm tra có khớp không
        if (ch == ')' and top == '(') or \
           (ch == ']' and top == '[') or \
           (ch == '}' and top == '{'):

            stack.pop()

        else:
            valid = False
            break

# Nếu stack còn phần tử -> sai
if len(stack) != 0:
    valid = False

if valid:
    print("Hop le")
else:
    print("Khong hop le")
```

```python
N = int(input("Nhap N: "))

# Hàm kiểm tra số nguyên tố
def is_prime(n):

    if n < 2:
        return False

    for i in range(2, int(n ** 0.5) + 1):

        if n % i == 0:
            return False

    return True

max_length = 0
current = []

result = []

# Duyệt từ 1 đến N
for num in range(1, N + 1):

    # Nếu là hợp số
    if num > 1 and not is_prime(num):

        current.append(num)

    else:

        # Kiểm tra độ dài lớn nhất
        if len(current) > max_length:

            max_length = len(current)
            result = current.copy()

        current = []

# Kiểm tra lần cuối
if len(current) > max_length:

    result = current.copy()

print("Day hop so dai nhat:")
print(*result)
```

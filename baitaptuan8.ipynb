W7A16: Cho một mảng số nguyên không âm. Gọi
mảng đó là "mảng đẹp" nếu mỗi phần tử
x trong mảng xuất hiện đúng x lần.
Kiểm tra xem mảng có phải là mảng đẹp
hay không. Nếu không hãy xóa đi y số (y
là nhỏ nhất có thể) để mảng còn lại thành
mảng đẹp

```python
n = list(map(int,input().split()))
a = set(n)
dem = 0 
for i in a : 
    if i > n.count(i) :
        dem += n.count(i)
    elif i < n.count(i) : 
        dem += (n.count(i)-i)
if dem == 0 : 
    print("YES")
else : 
    print(f"NO {dem}")
```

W7A18: Sắp xếp có điều kiện: Số chẵn đứng
trước, số lẻ đứng sau, chẵn tăng dần, lẻ
giảm dần


```python
n = list(map(int,input().split()))
le = []
chan = []
for i in n : 
    if i % 2 == 0 : 
        chan.append(i)
    elif i % 2 != 0 : 
        le.append(i)
for a in range(len(le)-1) : 
    for b in range(a+1,len(le)) :
        if le[a] < le[b] : 
            le[a],le[b] = le[b],le[a]
            
for c in range(len(chan)-1) : 
    for d in range(c+1,len(chan)) :
        if chan[c] > chan[d] : 
            chan[c],chan[d] = chan[d],chan[c]
lst = chan + le 
print(*lst)            
```

W7A22 : Cho N tên sinh viên gồm cả họ và tên,
tìm xem tên nào hay gặp nhất. Nếu có
nhiều tên cùng số lần gặp thì lấy tên có
thứ tự từ điển nhỏ nhấ

```python
n = int(input())
ds = []
for i in range(n) : 
    a = input().split()
    b = a[-1]
    ds.append(b)
c = 0 
ten_nhieu_nhat = None
for ten1 in ds : 
    dem = 0 
    for ten2 in ds : 
        if ten1 == ten2 : 
            dem = dem + 1 
    if dem > c : 
        c = dem 
        ten_nhieu_nhat = ten1
    elif dem == c : 
        if ten1 < ten_nhieu_nhat : 
            ten_nhieu_nhat = ten1 
print(ds)
print(ten_nhieu_nhat)
```

W7A26: Tách các số có trong một đoạn văn bản,
tính tổng và tính trung bình các số đó. 

```python
n = input()
text = ""
for i in n : 
    if i.isdigit() : 
        text += i 
    else :
        text += " "
a = text.split()
lst = []
for j in a : 
    lst.append(int(j))
tong = 0 
dem = 0 
for so in lst : 
    tong += so 
    dem += 1
tb = tong / dem
if tb.is_integer() : 
    print(tong,int(tb))
else : 
    print(tong,tb)


```

W7A27: Cho một dãy ngoặc ()(){}{}[][]..., kiểm tra
xem dãy ngoặc đã cho có hợp lệ hay
không

```python
n = input()
while "()" in n or "[]" in n or "{}" in n : 
    n = n.replace("()","")
    n = n.replace("[]","")
    n = n.replace("{}","")
if len(n) == 0 : 
    print("YES")
else : 
    print("NO")
```

W7A30 : Tìm Dãy con tăng dần KHÔNG liên tục
dài nhất của một danh sách nhập vào từ
bàn phím

```python
n = list(map(int,input().split()))
a = []
for i in range(len(n)):
    b = [n[i]]
    for j in range(i+1,len(n)):
        if n[j] >= b[-1] : 
            b.append(n[j])
    if len(b) > len(a) : 
        a = b
print(*a)
```

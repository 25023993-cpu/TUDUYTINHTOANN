W5A3 : Viết hàm kiểm tra một số có phải là số nguyên tố hay không?

```python
def so_nguyen_to (n) : 
    if n < 2 : 
        return False 
    for i in range(2,n) : 
        if n % i == 0 : 
            return False 
    return True 
n = int(input())        
print(so_nguyen_to (n))

```

W5A4 : VIết hàm kiểm tra một số có phải số hoàn hảo hay không, xử lý trên dữ liệu kiểu số nguyên

```python
def so_hoan_hao(n) : 
    tong = 0 
    for i in range(1,n) : 
        if n % i == 0 : 
            tong = tong + i 
            if tong == n : 
                return True 
    return False 
n = int(input())
print(so_hoan_hao(n))       
```

W5A8 : Cho hai số nguyên x và y, hãy viết hàm trả về khoảng cách Hamming giữa hai số này

```python
def khoang_cach(x,y) : 
    n = x^y 
    so = 0 
    while n > 0  : 
        if n & 1 : 
         so = so + 1 
        n= n >> 1 
    return so 
x , y = map(int,input().split())
print(khoang_cach(x,y))
    
```

W5A9 : Viết hàm trả về tổng các chữ số của số nguyên dương n

```python
def tong(n) : 
    tong = 0 
    while n > 0 :  
        x = n % 10 
        tong = tong + x 
        n = n // 10 
    return tong 
n = int(input())
print(tong(n))
```

W5A10 : Hai từ được gọi là đẳng cấu nếu các kí tự trong một từ có thể được ánh xạ để được từ còn lại. Ánh
xạ một kí tự nghĩa là thay đổi toàn bộ các kí tự đó trong từ bởi một kí tự khác. Thứ tự của các kí tự
được giữ không đổi. Hai kí tự khác nhau không thể ánh xạ đến cùng một kí tự, nhưng một kí tự có
thể ánh xạ thành chính nó. Ví dụ, từ "abca" và "zbxz" là hai từ đẳng cấu vì ta có thể ánh xạ 'a' thành
'z', 'b' thành 'b' và 'c' thành 'x'.
Viết hàm trả về \(true\) nếu hai từ \(a\) và \(b\) là đẳng cấu, ngược lại \(false\). Bạn có thể giả sử hai
từ đầu vào có cùng độ dà

```python
def dang_cau(a,b) : 
    if len(a) != len(b) : 
        return False 
    for i in range(len(a)) : 
        for j in range (len(a)) : 
            if a[i] == a[j] : 
                if b[i] != b[j] : 
                    return False 
            if a[i] != a[j] :
                if b[i] == b[j] :
                    return False 
    return True
a , b = input().split()
print(dang_cau(a,b))
```

W5A12 : Viết hàm divide_with_remainder(dividend, divisor) trả về bộ (q, r) sao cho dividend = divisor*q + r và
0 <= |r| < |divisor|. 


```python
def divide_with_remainder(dividend, divisor) : 
    q = dividend // divisor  
    r = dividend % divisor 
    return q , r 
dividend , divisor = map(int,input().split())
print(divide_with_remainder(dividend, divisor))
```

W5A15 : Viết hàm is_right_triangle(a, b, c) trả về True nếu ba số có thể là ba cạnh của tam giác vuông.

```python
def is_right_triangle (a,b,c) : 
    if (a**2) == (b**2) + (c**2) or (b**2) == (a**2) + (c**2) or (c**2) == (a**2) +(b**2) : 
        return True 
    return False 
a , b , c = map(int,input().split())
print(is_right_triangle(a,b,c))
```
